## Project Submission: Continuous Control

## Learning algorithm

The submitted project solves the reacher environment by training a single agent. 

The environment was solved using a Deep Deterministic Policy Gradient algorithm. The agent was similar to the one used for the cartpole exercise. 

The agent uses two pairs of networks to learn the environment. Each pair consists of a local network and a target network identical to each other and is used for soft updates. 

The first pair corresponds to the Actor (or the Policy model). Each network has three fully connected layers. The first two layers consist of 128 nodes and use a relu activation function while the last uses an tanh activation function to provide a continuous action for the reacher. 

The second pair corresponds to the Critic (or the Value model). Each network here uses three fully connected layers as well with relu activation functions. The first two layers have 128 nodes each. The Critic maps the state, action pair to a Q-value. 

Both sets of networks are tuned using an Adam optimizer. A Ornstein-Uhlenbeck process is used to generate the noise. 

### The parameters used for the agent are:  
Minibatch size: 128  
Replay buffer size: 1000000  
Discount factor: 0.99 (GAMMA)  
Soft update factor: 0.001 (TAU)  
Learning rate of actor: 0.0002  
Learning rate of critic: 0.0002
